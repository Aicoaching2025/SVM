```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import svm

# Generate synthetic 2D data for two classes
np.random.seed(42)
X_class1 = np.random.randn(20, 2) + np.array([2, 2])
X_class2 = np.random.randn(20, 2) + np.array([-2, -2])
X = np.vstack((X_class1, X_class2))
y = np.array([0]*20 + [1]*20)

# Train SVM classifier with linear kernel
clf = svm.SVC(kernel='linear', C=1, probability=True)
clf.fit(X, y)

# Plot decision boundary and support vectors
plt.figure(figsize=(8, 5))
xx, yy = np.meshgrid(np.linspace(-6, 6, 500), np.linspace(-6, 6, 500))
Z = clf.decision_function(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

plt.contourf(xx, yy, Z, levels=np.linspace(Z.min(), 0, 7), cmap=plt.cm.PuBu)
plt.contourf(xx, yy, Z, levels=[0, Z.max()], colors='orange')
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='coolwarm', edgecolors='k')
plt.scatter(clf.support_vectors_[:, 0], clf.support_vectors_[:, 1], s=100, facecolors='none', edgecolors='k', label='Support Vectors')
plt.title('SVM Decision Boundary for Cancer Detection')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.show()
```
